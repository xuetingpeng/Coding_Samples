# Below codes parse the top 250 douban movies and return the movie name, link, douban score, and othere detials.

Douban is a community that Chinese Hipsers love to use. It provides music, movie, books informaiton with ratings and comments.

Here is the douban top 250 movie website:https://movie.douban.com/top250?start={}&filter=



```python
import requests
import time
from bs4 import BeautifulSoup

# put the url pattern
url = 'https://movie.douban.com/top250?start={}&filter='
urls = [url.format(n*25) for n in range (10)]

# design the fucntion
def douban(url):
  headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.87 Safari/537.36'
  }
  res = requests.get(url, headers = headers)
  soup=BeautifulSoup(res.text,'html.parser')
  items = soup.find_all('div',class_ = 'info') # here class need to be class_
  for i in items:
    #print(i)
    tag = i.find('a')
    name = tag.find(class_='title').text
    link = tag['href']
    
    score = i.find('span', class_ = 'rating_num').text
    details = i.find('p').text
    
    print('name:'+name + '\nlink:' + link + '\nscore:'+str(score) + '\ndetails:' + details.replace(' ','')+'\n\n')

# use for loop to parse the whole website
for i in urls:
  douban(i)
  time.sleep(1)
```
